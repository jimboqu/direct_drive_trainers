```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
turbo_path = "turbo/ddsheet.csv"
turbo_data = pd.read_csv(turbo_path)
print("Ok")
```

```python
turbo_data
```

```python
plt.figure(figsize=(30,12))
sns.regplot(x=turbo_data['price'], y=turbo_data['incline'])
sns.scatterplot(x=turbo_data['price'], y=turbo_data['incline'], size=turbo_data['Power'], sizes=(100, 2000), 
                hue=turbo_data['Model'])
```

```python
plt.figure(figsize=(30,12))
sns.regplot(x=turbo_data['price'], y=turbo_data['Accuracy'])
sns.scatterplot(x=turbo_data['price'], y=turbo_data['Accuracy'], size=turbo_data['incline'], sizes=(100, 2000), 
                hue=turbo_data['Model'])
```
